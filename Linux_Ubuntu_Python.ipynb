## **Linux**

Linux es un sistema operativo de código abierto basado en el núcleo de Linux. Fue creado por Linus Torvalds en 1991 y desde entonces ha evolucionado con la colaboración de miles de desarrolladores en todo el mundo. A diferencia de otros sistemas operativos, Linux se distribuye bajo una licencia de software libre, lo que significa que es gratuito y permite a los usuarios acceder y modificar su código fuente.

Una de las características principales de Linux es su arquitectura modular, que permite adaptarlo a diferentes tipos de dispositivos y sistemas, desde servidores y computadoras de escritorio hasta dispositivos móviles y sistemas embebidos. Linux se destaca por su estabilidad, seguridad y capacidad para funcionar eficientemente incluso en hardware más antiguo.

Linux ofrece una amplia gama de distribuciones, también conocidas como "distros", que son diferentes variantes del sistema operativo. Estas distros pueden variar en términos de apariencia, conjunto de aplicaciones preinstaladas y enfoque de uso. Algunas de las distribuciones de Linux más populares incluyen Ubuntu, Fedora, Debian y CentOS.

## **Ubuntu**

Ubuntu es una de las distribuciones de Linux más conocidas y ampliamente utilizadas. Fue lanzada por primera vez en 2004 y está basada en el sistema operativo Debian. La filosofía detrás de Ubuntu se centra en proporcionar una experiencia de usuario fácil de usar y accesible para usuarios de todos los niveles de habilidad.

Una de las características distintivas de Ubuntu es su enfoque en la facilidad de instalación y configuración. La instalación de Ubuntu es relativamente sencilla y el sistema operativo viene con una amplia gama de controladores y software preinstalados, lo que facilita el uso del sistema después de la instalación.

Ubuntu también se destaca por su enfoque en la comunidad y la colaboración. La empresa detrás de Ubuntu, Canonical, promueve el desarrollo abierto y anima a los usuarios a contribuir al proyecto. Además, Canonical ofrece soporte y actualizaciones de seguridad a largo plazo para versiones específicas de Ubuntu, lo que proporciona estabilidad y confiabilidad a los usuarios.

Ubuntu es una distribución específica de Linux, lo que significa que se basa en el núcleo de Linux y utiliza muchas de sus funcionalidades y características.

- Código abierto: Tanto Linux como Ubuntu son sistemas operativos de código abierto. Esto significa que su código fuente está disponible para que los usuarios lo estudien, modifiquen y distribuyan libremente. La naturaleza de código abierto fomenta la transparencia, la colaboración y la flexibilidad.

- Comunidad activa: Tanto Linux como Ubuntu cuentan con comunidades activas de usuarios y desarrolladores en todo el mundo. Estas comunidades contribuyen al desarrollo, soporte técnico y documentación de los sistemas operativos. Además, ambas comunidades fomentan la participación y el intercambio de conocimientos.

- Personalización y flexibilidad: Tanto Linux como Ubuntu ofrecen una gran flexibilidad y personalización. Los usuarios pueden adaptar el sistema operativo a sus necesidades y preferencias mediante la instalación de diferentes aplicaciones, escritorios, temas y ajustes.

- Seguridad y estabilidad: Tanto Linux como Ubuntu son conocidos por su enfoque en la seguridad y la estabilidad. La arquitectura del sistema operativo y la amplia revisión del código fuente por parte de la comunidad ayudan a identificar y solucionar rápidamente posibles vulnerabilidades y errores.

**Resumen**: Linux es un sistema operativo de código abierto basado en el núcleo de Linux, mientras que Ubuntu es una de las distribuciones más populares de Linux. Ambos comparten un enfoque en el código abierto, una comunidad activa, personalización, seguridad y estabilidad. Sin embargo, Ubuntu se destaca por su enfoque en la facilidad de uso y la accesibilidad para usuarios de todos los niveles de habilidad.

## **Editores de texto en ubuntu**

Ubuntu al ser un sistema operativo pensado para el desarrollo de código, precisa de algunas herramientas útiles para esto. Específicamente hablamos de los editores de texto. Un editor de texto es una herramienta de software que permite crear y modificar archivos de texto. Su función principal es facilitar la edición y manipulación del contenido de archivos de texto, como documentos, código fuente, scripts, notas, entre otros.

Los editores de texto proporcionan una interfaz donde los usuarios pueden escribir, insertar, eliminar y formatear texto. Además, suelen ofrecer características adicionales, como resaltado de sintaxis para diferentes lenguajes de programación, autocompletado de palabras, búsqueda y reemplazo de texto, numeración de líneas, administración de proyectos, herramientas de depuración, integración con sistemas de control de versiones y más.

Acontinuación se enlistan algunos de los principales editores de texto que pueden ser usados en  ubuntu.


1. **Gedit**: Gedit es un editor de texto de código abierto y de uso general que se encuentra disponible en varios sistemas operativos, incluyendo Linux y macOS. Es el editor de texto predeterminado en el entorno de escritorio GNOME, pero también se puede utilizar en otros entornos de escritorio.

  Gedit ofrece una interfaz de usuario simple y fácil de usar, lo que lo hace adecuado para usuarios principiantes y aquellos que buscan un editor de texto ligero. Aunque Gedit se enfoca principalmente en la edición de texto plano, también ofrece características adicionales que pueden ser útiles para programadores y desarrolladores, como resaltado de sintaxis para varios lenguajes de programación, numeración de líneas, pestañas de documentos y la posibilidad de ampliar su funcionalidad mediante complementos.

  Algunas de las características y funcionalidades clave de Gedit incluyen:

  - Interfaz de usuario intuitiva: Gedit tiene una interfaz de usuario limpia y sencilla, con una barra de menú, una barra de herramientas y un área de edición de texto principal. Es fácil de navegar y utilizar.

  - Resaltado de sintaxis: Gedit reconoce la estructura y la sintaxis de varios lenguajes de programación y resalta la sintaxis correspondiente para mejorar la legibilidad del código.

  - Búsqueda y reemplazo: Gedit ofrece opciones de búsqueda y reemplazo de texto, lo que permite buscar palabras o frases específicas en un documento y reemplazarlas según sea necesario.

  - Gestión de pestañas: Permite abrir varios documentos en pestañas separadas, lo que facilita la navegación y el trabajo con múltiples archivos al mismo tiempo.

  - Complementos y personalización: Gedit se puede ampliar con una variedad de complementos para agregar funcionalidad adicional, como la integración con herramientas externas o la capacidad de editar archivos remotos a través de FTP.

## **Jupyter**

Jupyter es un proyecto de código abierto que proporciona un entorno interactivo para la creación y uso de documentos llamados "notebooks". Un Jupyter Notebook es un documento web que permite combinar código, texto explicativo, imágenes, gráficos y otros elementos en un solo archivo.

El Jupyter Notebook se basa en la idea de celdas, donde cada celda puede contener tanto código como texto. Esto permite la creación de documentos interactivos en los que los usuarios pueden ejecutar código directamente en el navegador web y ver los resultados de forma inmediata.

Las principales características y ventajas de Jupyter y los Jupyter Notebooks son:

1. Flexibilidad: Los Jupyter Notebooks admiten múltiples lenguajes de programación, incluyendo Python, R, Julia y muchos otros. Esto permite a los usuarios trabajar con diferentes lenguajes en un mismo documento.

2. Interactividad: Los Jupyter Notebooks permiten la ejecución interactiva de código en tiempo real. Esto significa que los usuarios pueden ejecutar una celda de código y ver los resultados de inmediato, lo que facilita la exploración de datos y la experimentación.

3. Documentación enriquecida: Los Jupyter Notebooks permiten la combinación de texto explicativo, ecuaciones matemáticas, imágenes, gráficos y otros elementos visuales junto con el código. Esto facilita la creación de informes, tutoriales y presentaciones interactivas.

4. Compartir y colaborar: Los Jupyter Notebooks se pueden compartir fácilmente, lo que permite la colaboración entre diferentes usuarios. También es posible exportar los notebooks en diferentes formatos, como HTML, PDF o archivos Python, lo que facilita su distribución y visualización en diferentes plataformas.

5. Reproducibilidad: Al combinar el código y los resultados en un solo archivo, los Jupyter Notebooks permiten la reproducción de análisis y experimentos. Otros usuarios pueden ejecutar el notebook y obtener los mismos resultados, lo que mejora la transparencia y la reproducibilidad de los trabajos.

***REMARK:*** Es importante aclarar que jupyter es un entorno que en la mayoría de ocaciones se puede clasificar como "exploratorio". Un código en un jupyter notebook no debe ser pensado como un producto final para producción en entornos académicos de investigación ni industriales.

## **Lectura de archivos e ingreso de datos**

Acontinuación veremos algunos ejemplos sencillos de de entrada de datos. El primero: un archivo .txt o de texto plano

```python
l = [1, 2, 3]
print(l[1])
l[1] = 100
print(l[1])
```

```python
l1 = [1, 2, 3]
l2 = l1.copy()
print(l1[1])
l1[1] = 100
print(l1, l2)
```

```python
t = 1, 2, 3
#t = (1, 2, 3)
print(type(t))
print(t[1])
try:
  t[1] = 100
except Exception as e:
  print(e)

print(t[1])
```

Es importante aclarar que se ha usado la sentencia `with` con el objetivo de evitar la necesidad de usar la función `close`. En ese caso el documento permanece abierto mientras haya indentación del `with`. Una vez se desindente el archivo se cierra automáticamente.

En el caso anterior se lee el archivo completo, sin embargo en muchas ocasiones se puede desear leer lineas específicas o iterar a traves de ellas. Si ese es el caso

```python
#Nombre del archivo de entrada
archivo = 'texto.txt'

#abrir archivo de texto

with open(archivo, 'r') as t:
  #Extrayendo el contenido del archivo de entrada
  contenido = t.read()
  print(contenido)
```

Aquí se puede ver que hay más lineas en blanco de las que debería haber. En ese caso lo que se puede hacer es aprovecharnos de que `linea` es una variable de tipo `str` por lo que podemos aplicar el método `strip` de los strings. Este método permite eliminar los espacios extra al inicio y al final de una cadena de caracteres y esto incluye el salto de linea.

```python
#Nombre del archivo de entrada
archivo = 'texto.txt'

#abrir archivo de texto

with open(archivo, 'r') as t:
  for linea_texto in t:
    print(linea_texto)
```

```python
#Nombre del archivo de entrada
archivo = 'texto.txt'

#abrir archivo de texto

with open(archivo, 'r') as t:
  for linea_texto in t:
    print(linea_texto.strip())
```

Existen otros parámetros para para la función `open`

|Parámetro| Uso |
|---------|-----|
| `r` | Lectura |
| `w` | Escritura (modifica desde cero)|
| `a` | Escritura (modifica desde la última escritura del archivo) |

Veamos un par de ejemplos sencillos.

```python
frase = "Esta clase es muy tarde"

arch = "arch_2.txt"

with open(arch, 'w') as f :
  f.write(frase)
```

En este caso se creó el archivo `salida.txt` en la misma dirección en la que se encuentra la dirección en memoria del noteboook. Si quisiéramos modificarlo:

```python
frase_nueva = "Pero aún así es buena"

with open(arch, 'a') as p:
  p.write('\n' + frase_nueva)
```
